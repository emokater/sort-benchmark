```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Size": [],
    "Selection Sort": [],
    "Heap Sort": [],
    "Quick Sort": [],
    "Sort": []
}

with open("/Users/ekaterinagridneva/Desktop/hse/mp/sort-benchmark/info_time.txt", "r") as f:
    lines = f.readlines()
    size = 0
    for line in lines:
        if "Datasets" in line:
            size = int(line[8:-2])
            data["Size"].append(size)
        elif line[0] == "1":
            time = float(line[28:-1])
            data["Selection Sort"].append(time)
        elif line[0] == "2":
            time = float(line[26:-1])
            data["Heap Sort"].append(time)
        elif line[0] == "3":
            time = float(line[28:-1])
            data["Quick Sort"].append(time)
        elif line[0] == "4":
            time = float(line[14:-1])
            data["Sort"].append(time)

df = pd.DataFrame(data)

# общий
plt.figure(figsize=(10, 6))
plt.plot(df["Size"], df["Selection Sort"], label="Selection Sort", color="blue")
plt.plot(df["Size"], df["Heap Sort"], label="Heap Sort")
plt.plot(df["Size"], df["Quick Sort"], label="Quick Sort")
plt.plot(df["Size"], df["Sort"], label="Sort")
plt.xlabel("Dataset Size")
plt.ylabel("Time (sec)")
plt.title("All Sort")
plt.legend()
plt.grid(True)
plt.show()

# Отдельный график для Selection Sort
plt.figure(figsize=(10, 5))
plt.plot(data["Size"], data["Selection Sort"], label="Selection Sort", color="red")
plt.xlabel("Dataset Size")
plt.ylabel("Time (sec)")
plt.title("Only Selection Sort")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Отдельный график для остальных
plt.figure(figsize=(10, 5))
plt.plot(data["Size"], data["Heap Sort"], label="Heap Sort", color="green")
plt.plot(data["Size"], data["Quick Sort"], label="Quick Sort", color="blue")
plt.plot(data["Size"], data["Sort"], label="Sort", color="red")
plt.xlabel("Dataset Size")
plt.ylabel("Time (sec)")
plt.title("Others")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
